```python
import matplotlib.pyplot as plt
```

```python
x_data=[4.0391,1.3197,9.5613,0.5978,3.5316,0.1650,1.6899,7.3172]
y_data=[111.4215,10.0112,30.2991,1.0625,13.1776,-3.1976,6.7367,23.8550]
```

```python
W=tf.Variable(tf.random_uniform([1],-5.0,5.0))
b=tf.Variable(tf.random_uniform([1],-5.0,5.0))

X=tf.placeholder(tf.float32)
Y=tf.placeholder(tf.float32)
```

```python
hypo=W*X+b
cost=tf.reduce_mean(tf.square(hypo-Y))
```

```python
a=tf.Variable(0.01)
optimizer=tf.train.GradientDescentOptimizer(a)
train=optimizer.minimize(cost)

init=tf.global_variables_initializer()
sess=tf.Session()
sess.run(init)
```

```python
for step in range(2001):
    sess.run(train, feed_dict={X:x_data, Y: y_data})
    if step%20==0:
        print(step,sess.run(hypo, feed_dict={X:x_data}),sess.run(b))
```

```python
answer=sess.run(hypo, feed_dict={X:5})
print('when X=5, hypo= '+str(answer))
```

```python
plt.figure(1)
plt.title('Linear Regression')
plt.xlabel('x')
plt.ylabel('y')

#주어진 데이터들을 점으로 표시
plt.plot(x_data, y_data, 'ro')

#train하여 예측한 일차함수를 직선으로 표시
plt.plot(x_data, sess.run(W)*x_data+sess.run(b),'b')

#x=5일때의 계산 값
plt.plot([5],answer,'go')
plt.show()
```
